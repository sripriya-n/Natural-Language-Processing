```python
import numpy as np
import pandas as pd
import re
```

```python
df = pd.read_csv("tamil_sentiment_full_train.tsv", sep='\t')
```

```python
#df = df_all.sample(frac=1).reset_index(drop=True)
print(df)

#data set is restriced to 500
#df = df[:500]

df.head()
```

```python
X = df['text']
y = df['category']
X
```

```python
# Preprocessing
processed_tweets = []
 
for tweet in range(0, len(X)):  
    # Remove all the special characters
    processed_tweet = re.sub(r'\W', ' ', str(X[tweet]))
 
    # remove all single characters
    processed_tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ', processed_tweet)
 
    # Remove single characters from the start
    processed_tweet = re.sub(r'\^[a-zA-Z]\s+', ' ', processed_tweet) 
 
    # Substituting multiple spaces with single space
    processed_tweet= re.sub(r'\s+', ' ', processed_tweet, flags=re.I)
 
    # Removing prefixed 'b'
    processed_tweet = re.sub(r'^b\s+', '', processed_tweet)
 
    # Converting to Lowercase
    processed_tweet = processed_tweet.lower()
 
    processed_tweets.append(processed_tweet)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  
tfidfconverter = TfidfVectorizer(max_features=2000)  
X = tfidfconverter.fit_transform(processed_tweets).toarray()
print(X.shape)
```

```python
#from sklearn.model_selection import train_test_split  
#X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
X_train = X
y_train = y
```

```python
# Handling Validation data
#df_all = pd.read_csv("tamil_sentiment_full_dev.tsv", sep='\t')
#df = df_all.sample(frac=1).reset_index(drop=True)
#print(df)
```

```python
from sklearn.ensemble import RandomForestClassifier
text_classifier = RandomForestClassifier(n_estimators=100, random_state=0)  
text_classifier.fit(X_train, y_train)
 
```

```python
df_val = pd.read_csv("tamil_sentiment_full_dev.tsv", sep='\t')
print(df_val)
```

```python
X_val = df_val['text']
y_val = df_val['category']
# Preprocessing
val_processed_tweets = []
 
for tweet in range(0, len(X_val)):  
    # Remove all the special characters
    val_processed_tweet = re.sub(r'\W', ' ', str(X_val[tweet]))
 
    # remove all single characters
    val_processed_tweet = re.sub(r'\s+[a-zA-Z]\s+', ' ', val_processed_tweet)
 
    # Remove single characters from the start
    val_processed_tweet = re.sub(r'\^[a-zA-Z]\s+', ' ', val_processed_tweet) 
 
    # Substituting multiple spaces with single space
    val_processed_tweet= re.sub(r'\s+', ' ', val_processed_tweet, flags=re.I)
 
    # Removing prefixed 'b'
    val_processed_tweet = re.sub(r'^b\s+', '', val_processed_tweet)
 
    # Converting to Lowercase
    val_processed_tweet = val_processed_tweet.lower()
 
    val_processed_tweets.append(val_processed_tweet)
print(val_processed_tweets)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  
tfidfconv = TfidfVectorizer(max_features=2000)  
X_val = tfidfconv.fit_transform(val_processed_tweets).toarray()
print(X_val.shape)
```

```python
X_test = X_val
y_test = y_val
predictions = text_classifier.predict(X_test)
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
 
print(confusion_matrix(y_test,predictions))  
print(classification_report(y_test,predictions))  
print(accuracy_score(y_test, predictions))
```
